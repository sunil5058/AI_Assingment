6. Adding HDI Category Column:

```python
# Define a function to classify HDI
def classify_hdi(hdi):
    if hdi < 0.550:
        return "Low"
    elif 0.550 <= hdi <= 0.699:
        return "Medium"
    elif 0.700 <= hdi <= 0.799:
        return "High"
    elif hdi >= 0.800:
        return "Very High"
    else:
        return pd.NA

# Apply the function to create a new column
hdi_2022_df['HDI_Category'] = hdi_2022_df['hdi'].apply(classify_hdi)
hdi_2022_df[['country','hdi','HDI_Category']].head(5)
```

```python
#1. Create South Asia Subset:
south_asia=["Afghanistan", "Bangladesh", "Bhutan", "India",
"Maldives", "Nepal", "Pakistan", "Sri Lanka"]
sa_df=hdi_2022_df[hdi_2022_df['country'].isin(south_asia)]
sa_df.to_csv("HDI_SouthAsia.csv",index=False)
sa_df
```

```python
#2. Composite Development Score:
#Here, i took the data of countries from year 2022

#Composiste Score = 0.30 * Life Expectancy Index +0.30*GNI   per Capita Index

sa_df['Composite_Score'] = 0.30 * sa_df['life_expectancy'] + 0.30 * sa_df['gross_inc_percap']

#Rank South Asian countries based on Composite_Score

top5_df=sa_df.sort_values(by='Composite_Score',ascending=False).head(5)

```

```python
#Plot the top 5 countries in a horizontal bar chart.
plt.barh(top5_df['country'],top5_df['Composite_Score'],color='skyblue')
plt.xlabel('Composite Score')
plt.ylabel('Country')
plt.title('Top 5 South Asian Countries by Composite Score')
plt.gca().invert_yaxis()
plt.show()
```

```python
sa_df['HDI_Rank']=sa_df['hdi'].rank(ascending=False)
sa_df['Composite_Rank']=sa_df['Composite_Score'].rank(ascending=False)
sa_df[['country','Composite_Score','Composite_Rank','hdi','HDI_Rank']]
```

```python
sa_df['Rank_Difference']=sa_df['Composite_Rank']-sa_df['HDI_Rank']
sa_df[['country','Composite_Score','Composite_Rank','hdi','HDI_Rank','Rank_Difference']]
```

```python
# 3. Outlier Detection:
def detect_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1

    outliers=df[(df[column] < Q1 - 1.5 * IQR) | (df[column] > Q3 + 1.5 * IQR)]
    return outliers
hdi_outliers=detect_outliers(sa_df,'hdi')
gni_outliers=detect_outliers(sa_df,'gross_inc_percap')
print("HDI Outliers:")
print(hdi_outliers)
print("\nGNI Per Capita Outliers:")
print(gni_outliers)
```

```python
# 4. Exploring Metric Relationships:
gender_corr=sa_df['gender_development'].corr(sa_df['hdi'])
life_corr=sa_df['life_expectancy'].corr(sa_df['hdi'])
print("Correlation between Gender Development and HDI:", gender_corr)
print("Correlation between Life Expectancy and HDI:", life_corr)

```

```python
# Create scatter plots with trendlines to visualize the relationships.
sns.set(style="whitegrid")

# Gender Development vs HDI
sns.lmplot(x='gender_development', y='hdi', data=sa_df, height=6, aspect=1.2, scatter_kws={'s':50, 'color':'blue'}, line_kws={'color':'red'})
plt.xlabel("Gender Development Index")
plt.ylabel("Human Development Index (HDI)")
plt.title("Gender Development vs. HDI", fontsize=14)
plt.show()

# Life Expectancy vs HDI
sns.lmplot(x='life_expectancy', y='hdi', data=sa_df, height=6, aspect=1.2, scatter_kws={'s':50, 'color':'green'}, line_kws={'color':'red'})
plt.xlabel("Life Expectancy")
plt.ylabel("Human Development Index (HDI)")
plt.title("Life Expectancy vs. HDI", fontsize=14)
plt.show()
```

```python
#Discuss :
# - Which metric is most strongly related to HDI and shows the weakest relationship with HDI.
if abs(gender_corr)>abs(life_corr):
  print("Gender Development is most strongly related to HDI")
else:
  print("Life Expectancy is most strongly related to HDI")
```

```python
#Gap Analysis:
# GNI HDI Gap = "gross inc percap" − "hdi"
sa_df['GNI_HDI_Gap']=sa_df['gross_inc_percap']-sa_df['hdi']

# Rank South Asian countries by GNI HDI Gap in descending and ascending order.
dec_rank=sa_df.sort_values(by='GNI_HDI_Gap',ascending=False)
asc_rank=sa_df.sort_values(by='GNI_HDI_Gap',ascending=True)

print("Countries with Positive GNI HDI Gap:")
print(dec_rank[['country','GNI_HDI_Gap']].head(3))
print("\nCountries with Negative GNI HDI Gap:")
print(asc_rank[['country','GNI_HDI_Gap']].head(3))

```

```python
# Plot the top 3 positive gaps and top 3 negative gaps.
top_positive =dec_rank.head(3)
top_negative =asc_rank.head(3)
plt.figure(figsize=(10,6))
sns.barplot(x='country', y='GNI_HDI_Gap', data=top_positive, palette='Greens')
plt.xlabel('Country')
plt.ylabel('GNI HDI Gap')
plt.title('Top 3 Countries with Positive GNI HDI Gap')
plt.xticks(rotation=45)
plt.show()

sns.barplot(x='country', y='GNI_HDI_Gap', data=top_negative, palette='Reds')
plt.xlabel('Country')
plt.ylabel('GNI HDI Gap')
plt.title('Top 3 Countries with Negative GNI HDI Gap')
plt.xticks(rotation=45)
plt.show()
```

```python
# Identify which region performs better on average.
if sa_mean>me_mean:
  print("South Asia performs better on average")
else:
  print("Middle East performs better on average")
```

```python
# 3. Top and Bottom Performers:
# Using 2022 data only
sa_2022=sa_2020_2022_df[sa_2020_2022_df['year']==2022]
sa_2022['region'] = 'South Asia'

me_2022=me_2020_2022_df[me_2020_2022_df['year']==2022]
me_2022['region'] = 'Middle East'

# South Asia

sa_top3=sa_2022.sort_values(by='hdi',ascending=False).head(3)
sa_bottom3=sa_2022.sort_values(by='hdi',ascending=True).head(3)


# Middle East
me_top3=me_2022.sort_values(by='hdi',ascending=False).head(3)
me_bottom3=me_2022.sort_values(by='hdi',ascending=True).head(3)

print("Top 3 Countries in South Asia:")
print(sa_top3[['country','hdi']])

print("\nBottom 3 Countries in South Asia:")
print(sa_bottom3[['country','hdi']])

print("\nTop 3 Countries in Middle East:")
print(me_top3[['country','hdi']])

print("\nBottom 3 Countries in Middle East:")
print(me_bottom3[['country','hdi']])



```

```python
# Create scatter plots highlighting outliers in a different color.
def plot_outliers(df, column ,region_name):
  outliers=detect_outliers(df,column)

  plt.figure(figsize=(10,6))
  normal = df [~df['country'].isin(outliers['country'])]
  plt.scatter(df[column],df['hdi'],color='blue',label='Normal Data')

  if not outliers.empty:
    plt.scatter(outliers['country'],outliers[column],color='red',label='Outliers')

  plt.xticks(rotation=45)
  plt.ylabel(column)
  plt.title(f'{region_name}: {column} Outliers (2022)')
  plt.legend()
  plt.show()

  print(f"{region_name} {column} Outliers:")
  if outliers.empty:
    print("No outliers found.")
  else:
   print(outliers[['country',column]])

# South Asia
plot_outliers(sa_2022,'hdi','South Asia')
plot_outliers(sa_2022,'gross_inc_percap','South Asia')

# Middle East
plot_outliers(me_2022,'hdi','Middle East')
plot_outliers(me_2022,'gross_inc_percap','Middle East')


```
